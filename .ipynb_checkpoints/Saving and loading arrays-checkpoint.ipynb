
# Loading .npy files
hàm np.load(): tải dữ liệu tệp
Save arrays in many formats:
- .csv
- .txt
- .pkl
- .npy
## Đoạn mã dưới đây cung cấp thông tin như sau:
- mở tệp dưới dạng nhị phân ("rb") alias thành f
- tải dữ liệu tệp bằng hàm load
- dùng hàm imshow() hiển thị hình ảnh
- gọi hàm show() để hiển thị hình ảnh trong 1 cửa sổ

```python
import numpy as np
import matplotlib.pyplot as plt

with open("C:\\Users\\HI.WELCOME TO NET\\Downloads\\rgb_array.npy", "rb") as f:
	logo_rgb_array = np.load(f)
plt.imshow(logo_rgb_array)
plt.show()
```

```python
red_array = logo_rgb_array[:, :, 0]
blue_array = logo_rgb_array[:, :, 1]
green_array = logo_rgb_array[:, :, 2]
```

# Exammining RGB data(kiểm tra dữ liệu RGB)

```python
red_array[1], green_array[1], blue_array[1]
```

```python
dark_logo_array = np.where(logo_rgb_array == 255, 50, logo_rgb_array)
plt.imshow(dark_logo_array)

plt.show()
```

```python
with open("C:\\Users\\HI.WELCOME TO NET\\Downloads\\rgb_array.npy","wb") as f:
    np.save(f, dark_logo_array) # save file
```

# hàm flip() để đảo ngược mảng

```python
flipped_logo = np.flip(logo_rgb_array)
plt.imshow(flipped_logo)
plt.show()
```

```python
flipped_colors_logo = np.flip(logo_rgb_array, axis=2)
plt.imshow(flipped_colors_logo)
plt.show()
```

Từ thư viện NumPy để đảo ngược mảng logo_rgb_array theo trục thứ ba (axis=2).
Điều này đảo ngược các thành phần màu (R, G, B) của mỗi điểm ảnh trong mảng.
