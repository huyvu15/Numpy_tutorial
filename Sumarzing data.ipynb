# Aggregating methods

- sum()

- min()

- max()

- mean()

- cumsum()

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([[0, 5, 1],
[0, 2, 0],
[1, 1, 2],
[2, 2, 1],
[0, 0, 0]])
a
```

```python
# Sum
a.sum()
```

```python
# caculate row, colomns
a.sum(axis = 0)
a.sum(axis = 1)
```

```python
# max, min
a.max()
```

```python
a.min()
```

```python
a.min(axis = 1)
```

```python
# trung bình
a.mean()
```

```python
# cumsum: tính tổng tích lũy 
a.cumsum(axis=0)
```

# Graphing summary values

```python
b = a.cumsum(axis = 0)
plt.plot(np.arange(1, 6), a[:, 0], label = "x")
plt.plot(np.arange(1, 6), a.mean(axis = 1), label = "Averrage")
plt.legend()
plt.show()
```

# Plus, Multiplying(nhân) matrix with 3: 
don't use array = array + 3

```python
array = np.array([[1, 2, 3], [4, 5, 6]])
for row in range(array.shape[0]):
	for column in range(array.shape[1]):
		array[row][column] += 3
array
```

```python
# cộng trực tiếp, nhân cũng tương tự
array_a = np.array([[1, 2, 3], [4, 5, 6]])

array_b = np.array([[0, 1, 0], [1, 0, 1]])

array_a + array_b
```
